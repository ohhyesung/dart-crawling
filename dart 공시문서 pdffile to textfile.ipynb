# pdf file 에서 text extraction 하기

```python
#pdf file -> text extraction
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO

def convert_pdf_to_txt():
    #pdf리소스 매니저 객체 생성
    rsrcmgr = PDFResourceManager()
    #문자열 데이터를 파일처럼 처리하는 stringio ->pdf파일 내용 담김
    retstr = StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr,retstr,codec=codec, laparams = laparams)
    fp = open(r'C:\Users\hyesung\Downloads\5860150.pdf','rb')
    interpreter = PDFPageInterpreter(rsrcmgr,device)
    password=""
    maxpages = 100000
    caching = True
    pagenos = set()

    for page in PDFPage.get_pages(fp,pagenos, maxpages = maxpages, password = password, caching = True):
        interpreter.process_page(page)
    #text에 결과가 담김 
    text = retstr.getvalue()

    fp.close()
    device.close()
    retstr.close()
    return text
```

```python
v = convert_pdf_to_txt()
```
